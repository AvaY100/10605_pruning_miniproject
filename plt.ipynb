## Accelerate Inference: Neural Network Pruning

```python
import matplotlib.pyplot as plt

# Data for method 1: magnitude based weight pruning
sparsity_method_1 = [0.20, 0.30, 0.40, 0.50, 0.60, 0.70]
accuracy_method_1 = [0.7125, 0.6364, 0.5509, 0.4535, 0.2982, 0.2760]

# Data for method 2: magnitude based weight pruning with five more epochs of finetuning
sparsity_method_2 = [0.20, 0.30, 0.40, 0.50, 0.60, 0.70]
accuracy_method_2 = [0.7251, 0.7303, 0.7188, 0.6883, 0.6420, 0.6285]

# method 3: neuron pruning for dense pruning
sparsity_method_3 = [0.17895931269249474, 0.3561895499000378, 0.5334197871075809, 0.710650024315124]
accuracy_method_3 = [0.6435643434524536, 0.5247524976730347, 0.43920791149139404, 0.3378217816352844]

# Create plot
plt.figure(figsize=(10, 6))
plt.plot(sparsity_method_1, accuracy_method_1, label='Magnitude-based Weight Pruning', marker='o')
plt.plot(sparsity_method_2, accuracy_method_2, label='Magnitude-based Weight Pruning + Fine-tuning', marker='o')
plt.plot(sparsity_method_3, accuracy_method_3, label='Neuron Pruning', marker='o')



# Adding titles and labels
plt.title('Pareto Frontier of Accuracy vs. Sparsity')
plt.xlabel('Sparsity')
plt.ylabel('Accuracy')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
